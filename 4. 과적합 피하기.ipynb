```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import pandas as pd
import tensorflow as tf

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("dataset/sonar.csv", header=None)
```

```python
df.info()
```

```python
df.head()
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
dataset = df.values
X = dataset[:, :60].astype(np.float)
y_obj = dataset[:, 60]
```

```python
# 문자열 변환
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(y_obj)
y = le.transform(y_obj)
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(24, input_dim = 60, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss = 'binary_crossentropy',
             optimizer = 'adam',
             metrics = ['accuracy'])
```

```python
# 모델 실행
model.fit(X, y, epochs = 200, batch_size = 5)
```

```python
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, y)[1]))
```

```python
# 과적합 방지를 위해 Hold-out 검증
from sklearn.model_selection import train_test_split

# 학습셋과 테스트셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.3)
```

```python
model.fit(X_train, Y_train, epochs = 130, batch_size = 5)
```

```python
# 테스트셋에 모델 적용
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```

```python
from sklearn.model_selection import StratifiedKFold

n_fold = 10
skf = StratifiedKFold(n_splits = n_fold, shuffle = True, random_state = 10)
```

```python
accuracy = []

for idx, (train, test) in enumerate(skf.split(X, y)):
    model = Sequential()
    model.add(Dense(24, input_dim = 60, activation = 'relu'))
    model.add(Dense(10, activation = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss = 'mean_squared_error',
                 optimizer = 'adam',
                 metrics = ['accuracy'])
    
    model.fit(X[train], y[train], epochs = 100, batch_size = 5)
    k_accuracy = (model.evaluate(X[test], y[test]))[1]
    
    print("folder : %d, accuracy : %.4f" %(idx, k_accuracy))
    
    accuracy.append(k_accuracy)
    
print("\n Test accuracy :", accuracy)
print("\n accuracy : %.4f" %(np.mean(accuracy)))
    
```
